```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Pre-Processing

```python
titanic_data = pd.read_csv('train.csv')
```

Handling the Missing Values

```python
# drop cabin column
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
```

```python
# replacing missing values in 'Age' column with mean value
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
# finding the mode value of 'Embarked' column
print(titanic_data['Embarked'].mode())
```

```python
# replacing the missing values in 'Embarked' column with mode value
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
titanic_data.isnull().sum()
```

Encoding the Categorical Columns 

```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
titanic_data.replace({'Sex': {'male':0, 'female':1},
                      'Embarked': {'S':0, 'C':1, 'Q':2}},
                     inplace=True)
```

```python
titanic_data.head()
```

Splitting into Target & Feature

```python
X = titanic_data.drop(columns=['Name', 'Ticket', 'PassengerId', 'Survived'], axis=1)
Y = titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

Splitting into Training & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training using Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
print(X_train_prediction)
training_accuracy_score = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on Training Data:- ', training_accuracy_score)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
print(X_test_prediction)
test_accuracy_score = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on Test Data:- ', test_accuracy_score)
```
