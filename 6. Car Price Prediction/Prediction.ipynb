Regression Problem

Importing Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data Collection & Processing

```python
car_dataset = pd.read_csv('dataset.csv')
car_dataset.shape
```

```python
car_dataset.head()
```

```python
car_dataset.isnull().sum()
```

```python
car_dataset.info()
```

```python
# checking distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encoding the Categorical Data

```python
# convert categorical to numerical values
car_dataset.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2},
                     'Seller_Type':{'Dealer':0, 'Individual':1},
                     'Transmission':{'Manual':0, 'Automatic':1}},
                    inplace=True)
```

```python
car_dataset.head()
```

Splitting into data & target

```python
X = car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y = car_dataset['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

Splitting into training & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

1. Linear Regression

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train, Y_train)
```

Model Evaluation

```python
# prediction on training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print('R squared Error: ', error_score)
```

R² (R-squared) score quantifies how much better your regression line predicts compared to always predicting the mean value. 

0 means your model is no better than just predicting the mean. 1 means your model perfectly predicts all data points.

Higher R² indicates a better fit of your regression line to the actual data.

Visualizing actual & predicted prizes

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
# prediction on test data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print('R squared Error: ', error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

2. Lasso Regression

```python
las_reg_model = Lasso()
```

```python
las_reg_model.fit(X_train, Y_train)
```

```python
training_data_prediction = las_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction)
print('R squared Error: ', error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

```python
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data_prediction)
print('R squared Error: ', error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
